```python
#data
#y = data['V1']
X = data.ix[:, 'V1':]
```

```python
X, y = make_blobs(n_samples=5100, centers=3, n_features=36, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
plt.scatter(X[y==1]['V1'], X[y==1]['V2'], label='V1', c='red') 
plt.scatter(X[y==2]['V1'], X[y==2]['V2'], label='V2', c='blue')
plt.scatter(X[y==3]['V1'], X[y==3]['V2'], label='V3', c='lightgreen')

plt.legend()
plt.xlabel('V1')
plt.ylabel('V2')
plt.axis([0, 200, 0, 200])
plt.show()
```
